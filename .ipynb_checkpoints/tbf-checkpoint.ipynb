```python
state_params = 'eJwljDkKgEAMAL8iqQ1s3FOfIX7ArEllsaiV4t8l2M3AMA-cMHXAzrtx8wEHKoREIriGUDHqKJSpqipD38Fu8XXgMpvdZo2bsRoXHyUlV1GiG_4NR1Vk9lwoaQ5M8H6XZR1Y'

def get_data(state_params):

  url = "https://eapi.web.prod.cloud.atriumsports.com/v1/embed/106/fixture_detail"
  params = {
      'state': state_params
  }
  response = requests.get(url, params=params)

  data = response.json()

  return data

def get_match_data(data):

  competitors = data['data']['banner']['fixture']['competitors']
  df_competitors = pd.DataFrame(competitors)
  df_competitors = df_competitors.drop(columns=['secondaryScore', 'shootOutAttempts', 'resultStatus', 'link', 'draw'])

  data = data['data']['pbp']['4']['events']

  df_pbp = pd.DataFrame(data)

  def team_data_config():
    df_play_by_play = df_pbp.copy()
    df_teams = df_competitors.copy()

    team_data_df = df_play_by_play.groupby('entityId')['name'].apply(list).reset_index(name='ROSTER')

    team_data_df['TEAM_NAME'] = df_teams['name']
    team_data_df['TEAM_ROSTER_ID'] = df_teams['code'] + '_ROSTER'
    team_data_df['TEAM_CODE'] = df_teams['code']
    team_data_df['TEAM_HOME'] = df_teams['isHome']
    team_data_df['TEAM_LOGO'] = df_teams['logo']
    team_data_df['TEAM_COLOR'] = df_teams['colors']

    # team_data_df['TEAM_WON'] = df_teams['resultPlace'].apply(lambda x: x - 2 if x == 2 else x)
    team_data_df['TEAM_WON'] = df_teams['resultPlace'].apply(lambda x: x != 2).astype(int)
    team_data_df['TEAM_SCORE'] = df_teams['score']

    return team_data_df

  def dataframe_out():
    team_data = team_data_config()

    df_play_by_play = df_pbp.copy()

    # columns_to_merge = team_data.columns[~team_data.columns.isin(['ROSTER'])]
    cols = team_data.columns[team_data.columns.isin(['entityId', 'TEAM_ROSTER_ID', 'TEAM_NAME', 'TEAM_CODE', 'TEAM_HOME'])]

    df_play_by_play = pd.merge(df_play_by_play, team_data[cols], on='entityId', how='left')

    return df_play_by_play

  play_by_play, team_data = dataframe_out(), team_data_config()


  return play_by_play, team_data
```

```python
def play_by_play_config(pbp, teams):
    df_pbp = pbp.copy()
    df_teams = teams.copy()


    # Score Configurations
    team_home_id = df_teams.loc[df_teams['TEAM_HOME'], 'entityId'].values[0]
    team_away_id = df_teams.loc[~df_teams['TEAM_HOME'], 'entityId'].values[0]

    team_home_score = []
    team_away_score = []

    for scores in df_pbp['scores']:
        home_score = scores.get(team_home_id, 0)
        away_score = scores.get(team_away_id, 0)

        team_home_score.append(home_score)
        team_away_score.append(away_score)


    df_pbp['SCORE_HOME'] = team_home_score
    df_pbp['SCORE_AWAY'] = team_away_score

    # Time Configurations
    pattern = r'PT(\d+)M(\d+)S'

    # Extract minutes and seconds
    df_pbp[['minutes', 'seconds']] = df_pbp['clock'].str.extract(pattern).astype(int)

    # Convert to a float representing total minutes (e.g., 1 minute and 30 seconds becomes 1.50)
    df_pbp['time_decimal'] = df_pbp['minutes'] + df_pbp['seconds'] / 100.0

    # Convert to a timedelta format (e.g., 1 minute and 30 seconds becomes 00:01:30)
    # df_pbp['time_timedelta'] = pd.to_timedelta(df_pbp['minutes'], unit='m') + pd.to_timedelta(df_pbp['seconds'], unit='s')



    # Shooting Configurations
    df_shooting = df_pbp.copy()

    # shooting_cols = df_shooting.columns[df_shooting.columns.isin(['entityId', 'desc', 'eventId', 'eventSubType', 'eventType',  'success', 'x', 'y', 'successString'])]
    shooting_cols = df_shooting.columns[~df_shooting.columns.isin(['clock','personnImage', 'scores', 'minutes', 'seconds'])]

    df_shooting = df_shooting[shooting_cols]
    df_shooting = df_shooting.dropna()



    # Drop Columns
    df_pbp = df_pbp.drop(columns=['success', 'x', 'y', 'scores', 'personImage', 'minutes', 'secods'], axis=1)

    return df_pbp, df_shooting
```
